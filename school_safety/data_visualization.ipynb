```python
import pandas as pd

df_2019 = pd.read_csv("./2019_school_safety.csv")
df_2020 = pd.read_csv("./2020_school_safety.csv")
df_2021 = pd.read_csv("./2021_school_safety.csv")
df_2022 = pd.read_csv("./2022_school_safety.csv")
df_2023 = pd.read_csv("./2023_school_safety.csv")
```

```python
df = pd.concat([df_2019,df_2020,df_2021,df_2022,df_2023])
# df.info()
# df.head()
```

```python
df = df[df['학교급']=='초등학교']
# df.head()
```

```python
from matplotlib import font_manager, rc
import matplotlib.pyplot as plt

font_path = '/usr/share/fonts/truetype/nanum/NanumGothic.ttf'  # 설치된 NanumGothic 폰트 경로
font_manager.fontManager.addfont(font_path)
plt.rcParams['font.family'] = 'NanumGothic'
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 막대그래프 그리기, 데이터는 df, x축은 사고장소로 설정하여 사고장소별 사건 발생수를 나타낸 막대그래프를 그림
sns.countplot(data=df, x='사고장소')
# 그래프의 제목을 설정
plt.title("초등학교 사고장소별 사고발생 수")
# 그래프의 x축 이름을 90로 회전
plt.xticks(rotation=90)
```
